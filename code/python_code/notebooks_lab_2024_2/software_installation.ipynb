<h1 id="instalacion-de-github">Instalación de Github </h1>

---


## Windows

1. Descarga el instalador de Github

    [github.com](https://github.com/git-for-windows/git/releases/download/v2.46.0.windows.1/Git-2.46.0-64-bit.exe)

2. Ejecutar el instalador e instalar

    

3. Iniciar sesión desde la terminal usando las credenciales de su cuenta de Github. Para ello, es necesario abrir un Bash de Git:

    
   
    

    

    Ingrese los siguientes comandos con la información de su usuario de Github:
    
   
    <code>git config --global user.name "YOUR_USERNAME"</code>

    <code>git config --global user.email "YOUR_EMAIL"</code>


## Ubuntu

1. Antes de instalar cualquier software, actualiza el sistema operativo y los paquetes existentes. Abre una terminal y ejecuta los siguientes comandos:
   
    <code>sudo apt update</code>

    <code>sudo apt upgrade</code>


2. Instalar Git usando:

    <code>sudo apt install git-all</code>

1. Iniciar sesión desde una **NUEVA** terminal usando las credenciales de su cuenta de Github:
   
    <code>git config user.name "YOUR_USERNAME"</code>
    
    <code>git config --global user.email "YOUR_EMAIL"</code>

## Planta Motor DC
---

Conecte la planta del motor DC a un cargador USB o al computador. Está debería conectarse automáticamente a la red del laboratorio. Coloque en  <code>plant_number = "PC-M-XX"</code> el nombre de la planta que apareció en el puerto serial, la dirección del broker no la cambiamos.



```python
my_dcmotor = motor.MotorSystemIoT(plant_number = "PC-M-XX", broker_address = "192.168.1.100") 
```

Al ejecutar la siguiente celda, deberia obtenerse la respuesta al escalón con un controlador por defecto. Debido a ello, la rueda de la planta se movera y progresivamente se actualizará la grafica.

```python
motor.step_closed(my_dcmotor, t0=1,t1=2, r0=0,r1=180)
plt.show()
```

## Planta Térmica
---

Conecte la planta termica por USB a un cargador o al computador. Está debería conectarse automáticamente a la red del laboratorio.
Coloque en  <code>plant_number = "PC-T-XX"</code> el nombre de la planta que apareció en el puerto serial, la dirección del broker no la cambiamos.



```python
my_temp = temp.ThermalSystemIoT(plant_number = "PC-T-XX", broker_address = "192.168.1.100") 
```

Al ejecutar la siguiente celda, la planta debería dar la respuesta al escalón con un controladro que tiene por defecto. De esta manera, progresivamente se añadirán los puntos a la gráfica, así mismo, verá como lentamente cambiarán los colores de los leds de la planta indicando el valor de las señales.

```python
temp.step_closed(my_temp, t0=30,t1=30, r0=50,r1=60)
plt.show()
```
